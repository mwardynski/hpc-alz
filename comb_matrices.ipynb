```python
def calculate_and_print_new_roi_values(regional_errors, regional_error_multi, total_rois):
    rois = [0] * 166

    for cat in regional_errors:
        for roi_idx in range(len(regional_errors[cat])):
            rois[roi_idx] += regional_errors[cat][roi_idx] * regional_error_multi[cat]

    for roi_idx in range(len(rois)):
        rois[roi_idx] /= total_rois
        print("{:.19e}".format(rois[roi_idx]))
```
